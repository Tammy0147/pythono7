```python
import math
import random
numbers = []
counter = 1
while counter <= 5000:
    num = random.randint(1,2000)
    numbers.append(num)
    counter += 1
print(numbers) 
```

```python
rl= random.choice(Class)
SN= list(map(lambda rl:f'{random.choice(Class)}{random.choice(numbers)}',numbers))
print(SN)
```
